```python
set50name = pd.DataFrame(all_set50, columns=['set50'])
```

```python
set50name = set50name['set50'] + '.bk'
```

```python
set50name.values
```

```python
from pandas_datareader import DataReader
from datetime import datetime
close = DataReader(set50name.values, 'yahoo',start=datetime(2018,1,1))['Close']
close
```

```python
close_21 = close.rolling(window=21).mean().round(2)
close_200 = close.rolling(window=200).mean().round(2)
```

```python
# up
# 200 < 21 < price
pass_scan_more = []
for i in range(len(close.iloc[-1].index)):
    if close.iloc[-1].values[i] > close_21.iloc[-1].values[i] and close_21.iloc[-1].values[i] > close_200.iloc[-1].values[i]:
        pass_scan_more.append(close.iloc[-1].index[i])

print(len(pass_scan_more))          
pass_scan_more
```

```python
# down
# price < 21 < 200
pass_scan_less = []
for i in range(len(close.iloc[-1].index)):
    if close.iloc[-1].values[i] < close_21.iloc[-1].values[i] and close_21.iloc[-1].values[i] < close_200.iloc[-1].values[i]:
        pass_scan_less.append(close.iloc[-1].index[i])

print(len(pass_scan_less))          
pass_scan_less
```
