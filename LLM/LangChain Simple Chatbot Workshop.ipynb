```python
from langchain.chat_models import ChatOpenAI
from langchain.memory import ConversationBufferMemory
from langchain.chains import ConversationChain
from dotenv import load_dotenv
import os

# load environment variables
load_dotenv()

def create_chatbot():
    # สร้าง instance ของ ChatOpenAI
    llm = ChatOpenAI(
        temperature=0.7,
        model_name="gpt-3.5-turbo"
    )
    
    # สร้าง memory สำหรับเก็บประวัติการสนทนา
    memory = ConversationBufferMemory()
    
    # สร้าง conversation chain
    conversation = ConversationChain(
        llm=llm,
        memory=memory,
        verbose=True
    )
    
    return conversation

def chat_loop(conversation):
    print("Bot: สวัสดีครับ/ค่ะ! มีอะไรให้ช่วยไหมครับ/คะ? (พิมพ์ 'exit' เพื่อจบการสนทนา)")
    
    while True:
        user_input = input("You: ")
        
        if user_input.lower() == 'exit':
            print("Bot: ขอบคุณที่ใช้บริการครับ/ค่ะ!")
            break
            
        response = conversation.predict(input=user_input)
        print(f"Bot: {response}")

def main():
    # ตรวจสอบว่ามี API key หรือไม่
    if not os.getenv("OPENAI_API_KEY"):
        print("กรุณาตั้งค่า OPENAI_API_KEY ในไฟล์ .env ก่อน")
        return
        
    try:
        conversation = create_chatbot()
        chat_loop(conversation)
    except Exception as e:
        print(f"เกิดข้อผิดพลาด: {str(e)}")

if __name__ == "__main__":
    main()
```
